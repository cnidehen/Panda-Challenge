# School Summary

```python
* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results
```

```python
# Combine the data into one dataset.  
Summary_Complete = pd.merge(student_data, school_data, how="inner", on=["school_name", "school_name"])
school_summary_required = Summary_Complete.drop(["gender","Student ID","School ID"], axis=1)

# Checkthe dataframe created
# school_summary_required.head()

# Identify the school type 
Unique_Schools_Name_Group = school_summary_required.groupby("school_name")["type"].unique()
School_Type = pd.DataFrame({
    "school_name": Unique_Schools_Name_Group.index,
    "type": Unique_Schools_Name_Group.values
})
School_Type = School_Type.rename(columns={"type":"School Type"})

# Check the dataframe created
# School_Type

# Get the number of students for each school 
Schools_Name_Group_Count = school_summary_required.groupby("school_name")["student_name"].count()
School_StudentCount = pd.DataFrame({
    "school_name": Schools_Name_Group_Count.index,
    "student_name": Schools_Name_Group_Count.values
})
School_StudentCount = School_StudentCount.rename(columns={"student_name":"Total Students"})

# Check the dataframe created
# School_StudentCount

# Getting the total budget of each school 
School_Budget_Per = Summary_Complete.groupby("school_name")["budget"].unique()
School_Budget = pd.DataFrame({
    "school_name": School_Budget_Per.index,
    "budget": School_Budget_Per.values
})
School_Budget['budget'] = School_Budget['budget'].str.get(0)
School_Budget = School_Budget.rename(columns={"budget":"Total School Budget"})

# Check the dataframe created
# School_Budget

# Combine Dataframes
School_Type_StudentCount_Join = pd.merge(School_Type, School_StudentCount, how="inner", on=["school_name", "school_name"])
School_Type_StudentCount_Budget_Join = pd.merge(School_Type_StudentCount_Join, School_Budget, how="inner", on=["school_name", "school_name"])
School_T_C_B =School_Type_StudentCount_Budget_Join

# Check the dataframe created
# School_T_C_B

# Per Student Budget
School_T_C_B["Per Student Budget"] = School_T_C_B["Total School Budget"]/School_T_C_B["Total Students"]

# Checking the dataframe created
# School_T_C_B

# Average Maths Score
Schools_Avg_Math_Score = school_summary_required.groupby("school_name")["maths_score"].mean()
School_Avg_Math = pd.DataFrame({
    "school_name": Schools_Avg_Math_Score.index,
    "Average Maths Score": Schools_Avg_Math_Score.values
})
# School_Avg_Math
# Add average maths score to dataframe
School_T_C_B_AvgM = pd.merge(School_T_C_B, School_Avg_Math, how="inner", on=["school_name", "school_name"])

# Check the dataframe created
# School_T_C_B_AvgM

# Average Reading Score
Schools_Avg_Read_Score = school_summary_required.groupby("school_name")["reading_score"].mean()
Schools_Avg_Read_Score = pd.DataFrame({
    "school_name": Schools_Avg_Read_Score.index,
    "Average Reading Score": Schools_Avg_Read_Score.values
})

# Schools_Avg_Read_Score
# Add average reading score to dataframe
School_T_C_B_AvgM_AvgR = pd.merge(School_T_C_B_AvgM, Schools_Avg_Read_Score, how="inner", on=["school_name", "school_name"])

# Check the dataframe created
# School_T_C_B_AvgM_AvgR
```

```python
# Calculate the overall percentage of student that passed Maths for each School
# Calculte the number of students that passed Maths in each school
Math_Score_Over50 = school_summary_required.loc[school_summary_required["maths_score"] >= 50]
Pass_Maths_School = Math_Score_Over50.groupby("school_name")["maths_score"].count()

# Turn this into a dataframe that can be added to the existing dataframe
Pass_Maths_School = pd.DataFrame({
    "school_name": Pass_Maths_School.index,
    "Students Passing Maths": Pass_Maths_School.values})
Pass_Maths_School.head()

# Merge the dataframe created to the existing datafram
Pass_Maths_And_Summary = pd.merge(School_T_C_B_AvgM_AvgR, Pass_Maths_School, how="inner", on=["school_name", "school_name"])

# Calculate Percentage passing and add it as a coloumn
Pass_Maths_And_Summary["% Passing Maths"] = (Pass_Maths_And_Summary["Students Passing Maths"]\
                                             /Pass_Maths_And_Summary["Total Students"])*100
# Drop the count of students passing coloumn
Pass_Maths_And_Summary = Pass_Maths_And_Summary.drop(["Students Passing Maths"], axis=1)

# Check the dataframe created
# Pass_Maths_And_Summary
```

```python
# Calculate the overall % of student that passed Reading for each School
# Calculte the number of students that passed Maths in each school
Read_Score_Over50 = school_summary_required.loc[school_summary_required["reading_score"] >= 50]
Pass_Read_School = Read_Score_Over50.groupby("school_name")["reading_score"].count()

# create a dataframe that can be added to the existing dataframe
Pass_Read_School = pd.DataFrame({
    "school_name": Pass_Read_School.index,
    "Students Passing Reading": Pass_Read_School.values})
Pass_Read_School.head()

# Merge the dataframe created to the existing datafram
Pass_Read_And_Summary = pd.merge(Pass_Maths_And_Summary, Pass_Read_School, how="inner", on=["school_name", "school_name"])

# Calculate Percentage passing and add it as a coloumn
Pass_Read_And_Summary["% Passing Reading"] = (Pass_Read_And_Summary["Students Passing Reading"]\
                                             /Pass_Read_And_Summary["Total Students"])*100
# Drop the count of students passing coloumn
Pass_Read_And_Summary = Pass_Read_And_Summary.drop(["Students Passing Reading"], axis=1)

# Check for the dataframe created
# Pass_Read_And_Summary
```

```python
# Calculate the overall % of student that passed both reading and maths for each School
# Calculte the number of students that passed Maths in each school
school_summary_required["Overall Passing"] = (school_summary_required["maths_score"] >= 50)&(school_summary_required["reading_score"] >= 50)
Overall_Summary = school_summary_required.loc[school_summary_required["Overall Passing"] == True]
Overall_Pass_School = Overall_Summary.groupby("school_name")["Overall Passing"].count()

# Create this into a dataframe that can be added to the existing dataframe
Overall_Pass_School = pd.DataFrame({
    "school_name": Overall_Pass_School.index,
    "Students Overall Passing": Overall_Pass_School.values})

# Merge the dataframe created to the existing datafram
Pass_Overall_And_Summary = pd.merge(Pass_Read_And_Summary, Overall_Pass_School, how="inner", on=["school_name", "school_name"])

# Calculate Percentage number of students that passed and add it as a coloumn
Pass_Overall_And_Summary["% Overall Passing"] = (Pass_Overall_And_Summary["Students Overall Passing"]\
                                             /Pass_Overall_And_Summary["Total Students"])*100
# Drop the count of students passing coloumn
Pass_Overall_And_Summary = Pass_Overall_And_Summary.drop(["Students Overall Passing"], axis=1)

    # Check the dataframe created
    # Pass_Overall_And_Summary
```

```python
# Format Dataframe and Change Coloumn Headings
# Set index to School Name
Per_School_Summary = Pass_Overall_And_Summary.set_index("school_name")
Per_School_Summary

# Change formatting for the following columns "Total Students", "Total School Budget" and "Per Studnet Budget" required
Per_School_Summary["Total Students"] = Per_School_Summary["Total Students"].map("{:,}".format)
Per_School_Summary["Total School Budget"] = Per_School_Summary["Total School Budget"].map("${:,.2f}".format)
Per_School_Summary["Per Student Budget"] = Per_School_Summary["Per Student Budget"].map("${:,.2f}".format)
Per_School_Summary['School Type'] = Per_School_Summary['School Type'].astype(str).str.strip("[]").str.replace("'", "")

# Create variable holding dataframe for use in later analysis
School_Size_df = Per_School_Summary
School_Type_df = Per_School_Summary
School_Spending_df = Per_School_Summary

# Remove Column name from the index
Per_School_Summary.index.name = None

# Generate output of the school summary Dataframe
Per_School_Summary
```

# Scores by School Spending

```python
* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
```

```python
# Remove dollar signs from "Per Student Budget" column
School_Spending_df["Per Student Budget"] = School_Spending_df["Per Student Budget"].str.replace("$", "", regex=False)

# Convert "Per Student Budget" column to numeric type
School_Spending_df["Per Student Budget"] = School_Spending_df["Per Student Budget"].astype(float)

# Create bins in which to place values based upon per student budget
bins = [0, 585, 630, 645, 680]

# Create labels for these bins
group_labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Slice the data and place into the bins
School_Spending_df["Spending Ranges (Per Student)"] = pd.cut(School_Spending_df["Per Student Budget"], bins, labels = group_labels, include_lowest = True)
    #Checking the dataframe created
    #School_Spending_df

# Create the relevant groupby functions
spending_maths_scores = School_Spending_df.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["Average Maths Score"].map("{:.2f}".format)
spending_reading_scores = School_Spending_df.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["Average Reading Score"].map("{:.2f}".format)
spending_passing_maths = School_Spending_df.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["% Passing Maths"].map("{:.2f}".format)
spending_passing_reading = School_Spending_df.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["% Passing Reading"].map("{:.2f}".format)
overall_passing_spending = School_Spending_df.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["% Overall Passing"].map("{:.2f}".format)

# Construct the Dataframe based on relevant groupby functions
spending_summary_df = pd.DataFrame({
    "Average Maths Score": spending_maths_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Maths": spending_passing_maths,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})

spending_summary_df
```

```python
* Perform the same operations as above, based on school size.
```

```python

# Remove formating from "Total Students" column
School_Size_df["Total Students"] = School_Size_df["Total Students"].str.replace(",", "", regex=False)

# Convert "Per Student Budget" column to numeric type
School_Size_df["Total Students"] = School_Size_df["Total Students"].astype(float)

# Create bins in which to place values based upon student count 
bins = [0, 1000, 2000, 5000]

# Create labels for these bins
group_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Slice the data and place it into bins
School_Size_df["School Size"] = pd.cut(School_Size_df["Total Students"], bins, labels = group_labels)
    
        #Checking the dataframe created
        #School_Size_df

#Create the relevant groupby functions based on size
size_maths_scores = School_Size_df.groupby(["School Size"]).mean(numeric_only=True)["Average Maths Score"]
size_reading_scores = School_Size_df.groupby(["School Size"]).mean(numeric_only=True)["Average Reading Score"]
size_passing_maths = School_Size_df.groupby(["School Size"]).mean(numeric_only=True)["% Passing Maths"]
size_passing_reading = School_Size_df.groupby(["School Size"]).mean(numeric_only=True)["% Passing Reading"]
overall_passing_size = School_Size_df.groupby(["School Size"]).mean(numeric_only=True)["% Overall Passing"]

#Construct the Dataframe based on relevant groupby functions for School Size
Size_Summary = pd.DataFrame({
    "Average Maths Score": size_maths_scores,
    "Average Reading Score": size_reading_scores,
    "% Passing Maths": size_passing_maths,
    "% Passing Reading": size_passing_reading,
    "% Overall Passing": overall_passing_size
})

Size_Summary
```
